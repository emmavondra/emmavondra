```python
ne_nri = pd.read_csv("NRI_Table_CensusTracts_Nebraska.csv")
ne_svi = pd.read_csv("Nebraska.csv")
nri_dictionary = pd.read_csv("NRIDataDictionary.csv")
co_nri = pd.read_csv("NRI_Table_CensusTracts_Colorado.csv")
co_svi = pd.read_csv("Colorado.csv")

data = {'col1': [1, 2, np.nan, 4], 'col2': [5, np.nan, 7, 8]}
df = pd.DataFrame(data)

#fills na columns

df_filled_0 = df.fillna(0)
df_filled_mean = df.fillna(df.mean())

ne_nri_filled_0 = ne_nri.fillna(0)
ne_nri_filled_mean = ne_nri.fillna(df.mean())

co_nri_filled_0 = co_nri.fillna(0)
co_nri_filled_mean = co_nri.fillna(df.mean())

print(ne_svi)
print(co_svi)

```

```python
ne_nri_filled_0['STCNTY'] = ne_nri_filled_0['STCOFIPS']
co_nri_filled_0['STCNTY'] = co_nri_filled_0['STCOFIPS']
ne_nri_filled_0.columns
co_nri_filled_0.columns

complete_data_ne = pd.merge(ne_nri_filled_0, ne_svi, on = 'STCNTY')
complete_data_co = pd.merge(co_nri_filled_0, co_svi, on = 'STCNTY')

complete_data_ne.columns
complete_data_co.columns
```

```python
#merges the data
complete_data_ne = pd.merge(ne_nri_filled_0, ne_svi, on = 'STCNTY', suffixes=('', '_right'))
complete_data_ne.columns

complete_data_co = pd.merge(co_nri_filled_0, co_svi, on = 'STCNTY', suffixes=('', '_right'))
complete_data_co.columns

```

```python
#adding a new risk
complete_data_co['my_risk'] = complete_data_co['WFIR_EXP_AREA']*complete_data_co['WFIR_EXPT']
print(complete_data_co['my_risk'], complete_data_co['WFIR_EXP_AREA'], complete_data_co['WFIR_EXPT'])

complete_data_ne['my_risk'] = complete_data_ne['WFIR_EXP_AREA']*complete_data_ne['WFIR_EXPT']
print(complete_data_ne['my_risk'], complete_data_ne['WFIR_EXP_AREA'], complete_data_ne['WFIR_EXPT'])


```

```python
#new risk: Colorado
co_svi = co_svi[['FIPS', 'COUNTY']]

complete_data_co = co_nri_filled_0.merge(co_svi[['COUNTY']], on='COUNTY', how='left')

print(complete_data_co.columns.tolist())

complete_data_co['my_risk'] = complete_data_co['WFIR_EXP_AREA'] * complete_data_co['WFIR_EXPT']

grouped_data_co = complete_data_co.groupby('COUNTY').agg(
    mean_wildfire=('my_risk', 'mean'),
)

print(grouped_data_co)

```

```python
#new risk:nebraska
ne_svi = ne_svi[['FIPS', 'COUNTY']]

complete_data_ne = ne_nri_filled_0.merge(ne_svi[['COUNTY']], on='COUNTY', how='left')

print(complete_data_ne.columns.tolist())

complete_data_ne['my_risk'] = complete_data_ne['WFIR_EXP_AREA'] * complete_data_ne['WFIR_EXPT']

grouped_data_ne = complete_data_ne.groupby('COUNTY').agg(
    mean_wildfire=('my_risk', 'mean'),
)

print(grouped_data_ne)

```

```python
grouped_data_ne.plot(kind='bar', figsize=(20, 10), width=0.8)

# Customizing the plot
plt.xlabel('COUNTY')
plt.ylabel('my_risk')
plt.title('Wildfire Risk')
plt.xticks(rotation=90)

# Show the plot
plt.legend(title="Risk Type")  # Adding a legend to distinguish risk types
plt.show()

```
